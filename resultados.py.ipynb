Utilizou-se duas abordagens de classificação para resolver o seguinte problema: dado uma letra, a qual fonte ela pertence? 
A primeira abordagem foi a de DecisionTreeClassifier, que utilizou os cinco maiores arquivos em sua completude e obteve os seguintes resultados:

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
```

```python
decisiontree_accuracy = [97.29, 97.54, 97.68, 97.77, 97.90]
decisiontree_datasize = [0.5, 0.4, 0.3, 0.2, 0.1]

plt.xlabel('Data Size')
plt.ylabel('Accuracy')

plt.title('Decision Tree')

plt.grid(True)

plt.scatter(decisiontree_datasize, decisiontree_accuracy)
plt.show()
```

Sobre essa abordagem, é interessante observar como a alteração de 40% do tamanho do dataset de treinamento não aumentou em 1% a acurácia do algoritmo, evidenciando que é interessante um número elevado de linhas para o treino do algoritmo.

A segunda abordagem utilizou o kNN Classifier, que utiliza os vizinhos mais pertos de um ponto em um gráfico de duas dimensões para classificar algum objeto específico. Surpreendentemente, nenhum hardware do grupo foi capaz de rodar todas as linhas dos cinco maiores arquivos. Por esse motivo, utilizou-se uma condição de 5000 linhas de cada arquivo. O resultado é descrito abaixo:

```python
# k = 4
knn4_accuracy = [91.48, 92.18, 92.80, 93.46, 93.76]
knn_datasize = [0.5, 0.4, 0.3, 0.2, 0.1]

knn5_accuracy = [91.25, 92.04, 92.88, 93.42, 93.92]
knn6_accuracy = [90.60, 91.55, 92.37, 93.00, 93.64]
knn7_accuracy = [90.32, 91.22, 92.01, 92.82, 93.44]

plt.xlabel('Data Size')
plt.ylabel('Accuracy')

plt.title('KNN')

plt.grid(True)

blue_patch = mpatches.Patch(color='blue', label='Neighbor = 4')

green_patch = mpatches.Patch(color='green', label='Neighbor = 5')

red_patch = mpatches.Patch(color='red', label='Neighbor = 6')

purple_patch = mpatches.Patch(color='purple', label='Neighbor = 7')
plt.legend(handles=[blue_patch, green_patch, red_patch, purple_patch])

plt.scatter(knn_datasize, knn4_accuracy, facecolor='blue')
plt.scatter(knn_datasize, knn5_accuracy, facecolor='green')
plt.scatter(knn_datasize, knn6_accuracy, facecolor='red')
plt.scatter(knn_datasize, knn7_accuracy, facecolor='purple')
plt.show()
```

É interessante notar como a quantidade K de vizinhos altera a acurácia do algoritmo, sendo que esse número é específico para cada dataset. Nos treinamentos, foi observado k=4 como um bom valor.
